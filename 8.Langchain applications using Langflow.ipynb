### Method 3: Python Requests

````python
import requests
import json

def query_langflow(flow_id: str, question: str, base_url: str = "http://127.0.0.1:7860"):
    """
    Send a query to a running Langflow instance.
    
    Args:
        flow_id: The unique ID of your Langflow project
        question: Your question or input
        base_url: Langflow server URL
    
    Returns:
        The model's response
    """
    endpoint = f"{base_url}/api/v1/run/{flow_id}"
    
    payload = {
        "input_value": question,
        "output_type": "chat",
        "input_type": "chat"
    }
    
    headers = {"Content-Type": "application/json"}
    
    response = requests.post(endpoint, json=payload, headers=headers)
    
    if response.status_code == 200:
        result = response.json()
        # Extract the text response from the nested structure
        try:
            text = result['outputs'][0]['outputs'][0]['results']['message']['data']['text']
            return text
        except (KeyError, IndexError):
            return result
    else:
        return f"Error: {response.status_code} - {response.text}"

# Example usage (uncomment when Langflow is running):
# response = query_langflow(
#     flow_id="a195037a-1cac-4f9d-9737-4613107b0374",
#     question="What are the best hiking trails in Switzerland?"
# )
# print(response)
````
