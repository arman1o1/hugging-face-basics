## ğŸ› ï¸ Part 4: Creating Custom Tools

Sometimes built-in tools aren't enough. The `@tool` decorator makes creating custom tools easy!

### Key Requirements for Custom Tools:

1. âœ… Use the `@tool` decorator
2. âœ… Include type hints for parameters
3. âœ… Write a detailed docstring (CRITICAL for agent understanding)
4. âœ… Return a meaningful result

````python
from smolagents import CodeAgent, LiteLLMModel, tool
import math

@tool
def calculate_bmi(weight_kg: float, height_m: float) -> str:
    """
    Calculate Body Mass Index (BMI) and provide health category.
    
    Args:
        weight_kg: Weight of the person in kilograms.
        height_m: Height of the person in meters.
    
    Returns:
        str: A string containing the BMI value and health category.
    """
    bmi = weight_kg / (height_m ** 2)
    
    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal weight"
    elif bmi < 30:
        category = "Overweight"
    else:
        category = "Obese"
    
    return f"BMI: {bmi:.1f} - Category: {category}"

print("âœ… BMI calculator tool created!")
````

````python
# Create an agent with our custom BMI tool
health_agent = CodeAgent(
    tools=[calculate_bmi],
    model=model
)

# Test the agent
bmi_response = health_agent.run(
    "Calculate the BMI for someone who weighs 75 kg and is 1.80 meters tall."
)

print("\n" + "="*60)
print("âš–ï¸ BMI CALCULATION:")
print("="*60)
print(bmi_response)
````

### ğŸ§ª Creative Example: A Unit Converter Tool

Let's create a more complex tool that handles multiple types of conversions:

````python
@tool
def convert_units(value: float, from_unit: str, to_unit: str) -> str:
    """
    Convert a value from one unit to another.
    
    Supports conversions between:
    - Length: km, m, cm, mm, mi, ft, in
    - Weight: kg, g, lb, oz
    - Temperature: celsius, fahrenheit, kelvin
    
    Args:
        value: The numeric value to convert.
        from_unit: The source unit (e.g., 'km', 'celsius', 'kg').
        to_unit: The target unit (e.g., 'mi', 'fahrenheit', 'lb').
    
    Returns:
        str: The conversion result with both values displayed.
    """
    # Conversion factors to base units
    length_to_meters = {
        'km': 1000, 'm': 1, 'cm': 0.01, 'mm': 0.001,
        'mi': 1609.344, 'ft': 0.3048, 'in': 0.0254
    }
    
    weight_to_kg = {
        'kg': 1, 'g': 0.001, 'lb': 0.453592, 'oz': 0.0283495
    }
    
    from_unit = from_unit.lower()
    to_unit = to_unit.lower()
    
    # Temperature conversions
    if from_unit in ['celsius', 'fahrenheit', 'kelvin']:
        if from_unit == 'celsius':
            celsius = value
        elif from_unit == 'fahrenheit':
            celsius = (value - 32) * 5/9
        else:  # kelvin
            celsius = value - 273.15
        
        if to_unit == 'celsius':
            result = celsius
        elif to_unit == 'fahrenheit':
            result = (celsius * 9/5) + 32
        else:  # kelvin
            result = celsius + 273.15
        
        return f"{value} {from_unit} = {result:.2f} {to_unit}"
    
    # Length conversions
    if from_unit in length_to_meters and to_unit in length_to_meters:
        meters = value * length_to_meters[from_unit]
        result = meters / length_to_meters[to_unit]
        return f"{value} {from_unit} = {result:.4f} {to_unit}"
    
    # Weight conversions
    if from_unit in weight_to_kg and to_unit in weight_to_kg:
        kg = value * weight_to_kg[from_unit]
        result = kg / weight_to_kg[to_unit]
        return f"{value} {from_unit} = {result:.4f} {to_unit}"
    
    return f"Cannot convert from {from_unit} to {to_unit}. Unsupported units."

print("âœ… Unit converter tool created!")
````

````python
# Create an agent with the unit converter
converter_agent = CodeAgent(
    tools=[convert_units],
    model=model
)

# Test various conversions
print("="*60)
print("ğŸ“ UNIT CONVERSIONS:")
print("="*60)

# Length conversion
response1 = converter_agent.run("Convert 5 kilometers to miles.")
print(f"\nğŸ›£ï¸ {response1}")

# Temperature conversion
response2 = converter_agent.run("Convert 98.6 fahrenheit to celsius.")
print(f"\nğŸŒ¡ï¸ {response2}")

# Weight conversion
response3 = converter_agent.run("Convert 150 pounds to kilograms.")
print(f"\nâš–ï¸ {response3}")
````

### ğŸ§ª Creative Example: A Text Analysis Tool

Let's create a tool for analyzing text content:

````python
import re
from collections import Counter

@tool
def analyze_text(text: str) -> str:
    """
    Analyze a piece of text and return various statistics.
    
    This tool provides comprehensive text analysis including:
    - Character count
    - Word count
    - Sentence count
    - Most common words
    - Average word length
    
    Args:
        text: The text string to analyze.
    
    Returns:
        str: A formatted report with all text statistics.
    """
    # Basic counts
    char_count = len(text)
    char_count_no_spaces = len(text.replace(' ', ''))
    
    # Word analysis
    words = re.findall(r'\b\w+\b', text.lower())
    word_count = len(words)
    
    # Sentence count (rough estimate)
    sentences = re.split(r'[.!?]+', text)
    sentence_count = len([s for s in sentences if s.strip()])
    
    # Most common words (excluding short ones)
    meaningful_words = [w for w in words if len(w) > 3]
    common_words = Counter(meaningful_words).most_common(5)
    
    # Average word length
    avg_word_len = sum(len(w) for w in words) / max(word_count, 1)
    
    # Format output
    report = f"""
ğŸ“Š TEXT ANALYSIS REPORT
{'='*30}
ğŸ“ Characters: {char_count} (without spaces: {char_count_no_spaces})
ğŸ“– Words: {word_count}
ğŸ“„ Sentences: {sentence_count}
ğŸ“ Avg Word Length: {avg_word_len:.1f} characters

ğŸ” Most Common Words:
"""
    for word, count in common_words:
        report += f"   â€¢ '{word}': {count} times\n"
    
    return report.strip()

print("âœ… Text analysis tool created!")
````

````python
# Create an agent with the text analysis tool
text_agent = CodeAgent(
    tools=[analyze_text],
    model=model
)

# Sample text for analysis
sample_text = """
Artificial intelligence is transforming the world. Machine learning algorithms 
are becoming more sophisticated every day. From healthcare to finance, AI is 
making significant impacts. The future of technology depends on how we develop 
and deploy these intelligent systems. Responsible AI development is crucial for 
ensuring positive outcomes for humanity.
"""

analysis_response = text_agent.run(
    f"Analyze this text: {sample_text}"
)

print(analysis_response)
````
